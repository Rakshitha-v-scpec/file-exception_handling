4.  Write a Python script that reads from one file and writes its content to another file.

```python
def copy_file(source_file, destination_file):
    try:
        with open(source_file, 'r') as src:
            content = src.read()

        with open(destination_file, 'w') as dest:
            dest.write(content)

        print(f"Content successfully copied from '{source_file}' to '{destination_file}'.")

    except FileNotFoundError:
        print(f"Error: The file '{source_file}' does not exist.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

# Example usage:
source = "input.txt"
destination = "output.txt"
copy_file(source, destination)
```

6.  Write a Python program that logs an error message to a log file when a division by zero exception occursF

```python
import logging

# Configure the logging system
logging.basicConfig(
    filename='error_log.log',      # Log file name
    level=logging.ERROR,           # Set minimum level to log
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def divide_numbers(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError as e:
        logging.error("Attempted division by zero: %s", e)
        print("Error: Division by zero is not allowed.")

# Example usage
divide_numbers(10, 0)
```

14. Write a program that uses the logging module to log both informational and error messages

```python
import logging

# Configure logging
logging.basicConfig(
    filename='app_log.log',          # Log file name
    level=logging.DEBUG,             # Capture all levels DEBUG and above
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def divide(a, b):
    logging.info(f"Attempting to divide {a} by {b}")
    try:
        result = a / b
        logging.info(f"Division successful: {a} / {b} = {result}")
        return result
    except ZeroDivisionError as e:
        logging.error(f"Error occurred: {e}")
        return None

# Example usage
divide(10, 2)     # Logs INFO messages
divide(5, 0)      # Logs ERROR message
```

18. How would you implement a basic logging setup that logs to a file with rotation after 1MB

```python
import logging
from logging.handlers import RotatingFileHandler

# Configure the rotating file handler
log_file = 'app_log.log'
max_log_size = 1 * 1024 * 1024  # 1MB in bytes
backup_count = 3  # Number of backup log files to keep

# Set up the logging configuration
handler = RotatingFileHandler(log_file, maxBytes=max_log_size, backupCount=backup_count)
handler.setLevel(logging.INFO)

# Create a formatter and set it for the handler
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Set up the logger
logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.addHandler(handler)

# Example usage of the logger
logger.info("This is an informational message.")
logger.error("This is an error message.")

# Simulate logging in a loop to generate log file content
for i in range(1000):
    logger.debug(f"Debug message #{i}")

```

21. Write a Python program that reads a file and prints the number of occurrences of a specific word

```python
def count_word_occurrences(filename, target_word):
    try:
        with open("F.txt", 'r') as file:
            content = file.read()
            # Normalize the text to lowercase for case-insensitive matching
            words = content.lower().split()
            count = words.count(target_word.lower())
            print(f"The word '{target_word}' occurs {count} time(s) in the file.")
    except FileNotFoundError:
        print(f"Error: The file does not exist.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

# Example usage
filename = "F.txt"
word_to_search = "line three"

count_word_occurrences(filename, word_to_search)
```
